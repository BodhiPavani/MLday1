```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_palette('husl')
import matplotlib as plt
%matplotlib inline
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('iris.csv')
data.head()
```

```python
data['Species'].value_counts()     #value_count gives no of count in the individual records
```

```python
tmp=data.drop('Id',axis=1)
g=sns.pairplot(tmp,hue='Species',markers='+')
```

```python
X=data.drop(['Id','Species'],axis=1)
Y=data['Species']
print(X.shape)
print(Y.shape)
```

```python
X.head()
```

```python
Y.head()    #we have seperated x features and y features
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
X_test
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
knn=KNeighborsClassifier(n_neighbors=12)
knn.fit(X_train,Y_train)
knn.predict([[2.1,8.9,5.6,4.0]])
```

```python
Y_pred=knn.predict(X_test)
print(metrics.accuracy_score(Y_test, Y_pred))
```

```python
data=pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
X=data.drop(['Pregnancies','BMI','DiabetesPedigreeFunction','Outcome'],axis=1)
Y=data['Outcome']
print(X.shape)
print(Y.shape)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
knn=KNeighborsClassifier(n_neighbors=12)
knn.fit(X_train,Y_train)
knn.predict([[147,80,30,230,32]])
```

```python
Y_pred=knn.predict(X_test)
print(metrics.accuracy_score(Y_test, Y_pred))
```
